```python
# Using wave open file in read mode
import wave
data_o = wave.open("/content/BabyElephantWalk60.wav",'r') # 2nd parameter 'r' specifies 'read'mode
```

```python
#display information about file
print( "Number of channels",data_o.getnchannels())
print( "Sample width",data_o.getsampwidth())
print( "Frame rate.",data_o.getframerate())
print("Number of frames",data_o.getnframes())
print( "parameters:",data_o.getparams())
```

```python
#read audio file
from scipy.io.wavfile import read
data_r = read("/content/BabyElephantWalk60.wav")

 #convert audio to array
import numpy as np
audio_to_arr = np.array(data_r[1],dtype=float)
```

```python
#type check
type(audio_to_arr)
```

```python
#convert array to dataframe
array_to_df = pd.DataFrame(audio_to_arr, columns=['A'])
```

```python
#print dataframe
print(array_to_df)
```

```python
#save dataframe to csv format
array_to_df.to_csv('audio_to_csv.csv')
```
